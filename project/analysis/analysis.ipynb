#### Instruction
1. Read the CSV file into a pandas dataframe. (1 pt)

```python
df = pd.read_csv('../data/Fremont_Bridge_Bicycle_Counter.csv')
df
```

2. Add columns to the dataframe containing: ( 3 pt)
  - i. The total (East + West) bicycle count (raw data already has this)
  - ii. The hour of the day
  - iii. The year

```python
# a helper method to convert string to a Timestamp object
def string_to_date(input_string):
    return pd.to_datetime(input_string, format='%m/%d/%Y %I:%M:%S %p')

# Add Hour and Year column to raw dataframe
df['Hour'] = df['Date'].apply(lambda x: float(string_to_date(x).hour))
df['Year'] = df['Date'].apply(lambda x: float(string_to_date(x).year))

# Check updated df
df.head()
```

3. Create a dataframe with the subset of data from the year 2016 (1 pt)

```python
df_2016 = df[df['Year'] == 2016].copy()
df_2016.head()
```

4. Use pandas + matplotlib to plot the counts by hour. (i.e. hour of the day on the x-axis, total daily counts on the y-axis) (1 pt)

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=df['Hour'], y=df['Fremont Bridge Total'])
plt.xlabel('hour of the day')
plt.ylabel('total daily counts')
plt.title('Counts by Hour')
plt.show()
```

5. Use pandas to determine what is (on average) the busiest hour of the day (1 pt)

```python
df_mean_total_by_hour = df.groupby(['Hour']).mean().sort_values(by=['Fremont Bridge Total'], ascending=False).copy()
df_mean_total_by_hour.head(1)
```

The busiest hour on average of the day is 17:00, 5pm.
